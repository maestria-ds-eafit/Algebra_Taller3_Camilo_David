```python
import matplotlib.pyplot as plt
```

```python
# Parámetros del modelo
beta = 0.1  # Tasa de transmisión
gamma = 0.15  # Tasa de recuperación
delta_t = 0.2  # Intervalo de tiempo

# Condiciones iniciales
S0 = 0.9  # Proporción inicial de susceptibles
I0 = 0.1  # Proporción inicial de infectados
R0 = 0.8  # Proporción inicial de recuperados

# Listas para almacenar los valores a lo largo del tiempo
S = [S0]
I = [I0]
R = [R0]

# Número total de pasos de tiempo
num_steps = 100

# Simulación del modelo SIR
for t in range(num_steps):
    dS = -beta * S[-1] * I[-1] * delta_t
    dI = (beta * S[-1] * I[-1] - gamma * I[-1]) * delta_t
    dR = gamma * I[-1] * delta_t
    
    S.append(S[-1] + dS)
    I.append(I[-1] + dI)
    R.append(R[-1] + dR)

# Visualización de los resultados
plt.figure(figsize=(10, 6))
plt.plot(range(num_steps + 1), S, label='Susceptibles')
plt.plot(range(num_steps + 1), I, label='Infectados')
plt.plot(range(num_steps + 1), R, label='Recuperados')
plt.xlabel('Días')
plt.ylabel('Proporción de la población')
plt.title('Modelo SIR')
plt.legend()
plt.show()
```
