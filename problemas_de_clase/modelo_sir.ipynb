```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P = np.array([[0.9, 0.1, 0], [0, 0.85, 0.15], [0.2, 0, 0.8]])
```

```python
data = np.array([0.5, 0.25, 0.25])
```

```python
x = []
y_susceptible = []
y_infected = []
y_recovered = []
projected_data = data
total = np.sum(data)

for i in range(100):
    x.append(i)
    susceptible_proportion = projected_data[0] / total
    infected_proportion = projected_data[1] / total
    recovered_proportion = projected_data[2] / total
    y_susceptible.append(susceptible_proportion)
    y_infected.append(infected_proportion)
    y_recovered.append(recovered_proportion)
    projected_data = P.T @ projected_data
```

```python
plt.plot(x, y_infected, label='Infected')
plt.plot(x, y_recovered, label='Recovered')
plt.plot(x, y_susceptible, label='Susceptible')
plt.legend()
plt.show()
```
